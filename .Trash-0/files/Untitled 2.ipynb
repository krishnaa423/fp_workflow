```python
a = np.linspace(-10, 10, 500)
x, y = np.meshgrid(a, a, indexing='ij')
z = np.sin(x**2 + y**2)

plt.style.use('bmh')
plt.pcolormesh(x.T, y.T, z.T, cmap='viridis')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interactive

# Define the function to plot
def plot_sine_wave(freq=1, amplitude=1):
    x = np.linspace(0, 2 * np.pi, 500)
    y = amplitude * np.sin(freq * x)
    plt.figure(figsize=(8, 4))
    plt.plot(x, y, label=f"Frequency: {freq}, Amplitude: {amplitude}")
    plt.ylim(-2, 2)
    plt.title("Interactive Sine Wave")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.legend()
    plt.grid(True)
    plt.show()

# Create interactive widgets
interactive_plot = interactive(
    plot_sine_wave, 
    freq=(0.1, 10, 0.1),  # Slider for frequency: min=0.1, max=10, step=0.1
    amplitude=(0.1, 2, 0.1)  # Slider for amplitude: min=0.1, max=2, step=0.1
)

# Display the interactive plot
interactive_plot
```

```python
from ipywidgets import interactive

a = np.linspace(-2, 2, 50)
x, y = np.meshgrid(a, a, indexing='ij')

def test_2d_interact(alpha):
    z = np.sin(alpha*(x**2 + y**2))
    plt.pcolormesh(x.T, y.T, z.T, cmap='viridis')
    plt.show()

interactive(test_2d_interact, alpha=(1, 5, 0.2))
```
