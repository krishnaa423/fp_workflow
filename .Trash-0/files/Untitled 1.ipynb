```r
# Create a simple data frame
data <- data.frame(
  Name = c("Alice", "Bob", "Charlie", "Diana"),
  Age = c(25, 30, 35, 40),
  Score = c(85, 90, 88, 92)
)

# Display the data
print(data)

# Summary statistics
summary(data)

# Basic plot
plot(data$Age, data$Score, main = "Age vs Score", xlab = "Age", ylab = "Score", pch = 19)
```
